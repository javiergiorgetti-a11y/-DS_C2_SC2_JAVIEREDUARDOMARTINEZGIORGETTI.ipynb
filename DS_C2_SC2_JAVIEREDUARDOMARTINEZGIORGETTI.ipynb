```python
# a) Extraer la información del archivo

import pandas as pd
df = pd.read_csv("country_vaccinations.csv")
df.head()
```

```python
# b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas,
#asegurándote que las columnas con fechas sean del tipo datetime64.

df.info()

df["date"] = pd.to_datetime(df["date"], errors="coerce")
df.info()
```

```python
#c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines,
#en otras palabras, agrupe por vaccines y realice la sumatoria).


df_sorted = df.sort_values(["country", "date"])
latest = df_sorted.groupby("country", as_index=False).tail(1)

vacunas_por_compania = (
    latest.groupby("vaccines", as_index=False)["total_vaccinations"]
    .sum()
    .sort_values("total_vaccinations", ascending=False)
)

vacunas_por_compania.head(10)
```

```python
# d. Obtener la cantidad de vacunas aplicadas en todo el mundo.
total_vacunas = latest["total_vaccinations"].sum()
total_vacunas
```

```python
# e. Calcular el promedio de vacunas aplicadas por país.

promedio_vacunas = latest["total_vaccinations"].mean()
promedio_vacunas
```

```python
# f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
fecha_obj = pd.to_datetime("2021-01-29")
total_vacunas_dia = df.loc[df["date"] == fecha_obj, "daily_vaccinations"].sum()
total_vacunas_dia
```

```python
#g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias)
#con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

conDiferencias = df.copy()

conDiferencias["daily_vaccinations"] = pd.to_numeric(conDiferencias["daily_vaccinations"], errors="coerce")
conDiferencias["daily_vaccinations_raw"] = pd.to_numeric(conDiferencias["daily_vaccinations_raw"], errors="coerce")

conDiferencias["diferencias"] = (
    conDiferencias["daily_vaccinations"] - conDiferencias["daily_vaccinations_raw"]
)

conDiferencias[["country","date","daily_vaccinations","daily_vaccinations_raw","diferencias"]].head()
```

```python
#h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

periodo = df["date"].max() - df["date"].min()
periodo
```

```python
# i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac)
#con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

conCantidad = df.copy()

conCantidad["canVac"] = (
    conCantidad["vaccines"]
    .fillna("")
    .apply(lambda x: len([v.strip() for v in x.split(",") if v.strip()]))
)

conCantidad[["country","date","vaccines","canVac"]].head()
```

```python
# j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

antes20 = df[df["date"] < pd.to_datetime("2020-12-20")].copy()
antes20.shape
```

```python
# k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

pfizer = df[df["vaccines"].fillna("").str.contains("Pfizer", case=False)].copy()
pfizer.shape

output = "resultadosReto.xlsx"
```

```python
# l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx,
#donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.


with pd.ExcelWriter(output, engine="openpyxl") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias", index=False)
    conCantidad.to_excel(writer, sheet_name="conCantidad", index=False)
    antes20.to_excel(writer, sheet_name="antes20", index=False)
    pfizer.to_excel(writer, sheet_name="pfizer", index=False)

output
```
